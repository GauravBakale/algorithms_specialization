### Dijkstra's shortest path

The file contains an adjacency list representation of an undirected weighted graph with 200 vertices labeled 1 to 200. Each row consists of the node tuples that are adjacent to that particular vertex along with the length of that edge. For example, the 6th row has 6 as the first entry indicating that this row corresponds to the vertex labeled 6. The next entry of this row "141,8200" indicates that there is an edge between vertex 6 and vertex 141 that has length 8200. The rest of the pairs of this row indicate the other vertices adjacent to vertex 6 and the lengths of the corresponding edges.

Your task is to run Dijkstra's shortest-path algorithm on this graph, using 1 (the first vertex) as the source vertex, and to compute the shortest-path distances between 1 and every other vertex of the graph. If there is no path between a vertex v and vertex 1, we'll define the shortest-path distance between 1 and v to be 1000000.

You should report the shortest-path distances to the following ten vertices, in order: 7,37,59,82,99,115,133,165,188,197. You should encode the distances as a comma-separated string of integers. Remember the order of reporting DOES MATTER, and the string should be in the same order in which the above ten vertices are given. The string should not contain any spaces. Dijkstras algorithm:


    Given a directed graph G, consider two lists X(nodes processed so far), (V-X)(nodes that are left). 
    
    Initialize X = [s] first vertex. 
    Then A[s] = 0 (computed shortest paths distance)
         B[s] = [] (computed shortest path)

    While X != V, among all the edges between the vertexes in X and (V-X) list, pick the shortest edge and add the head of this edge(lets say this is d) to the processed list X.
          In addition to this, 
               set A[d] = A[s] + l_sd and 
                   B[d] = B[s] + (s,d)

```python

def dijkstras_shortest_path(G: Graph, initial_node: Vertex):
    # initialise variables
    processed_nodes = []
    unprocessed_nodes = []
    shortest_length = {}
    shortest_path = {}

    # initial_node = G.vertexes['s']

    processed_nodes.append(initial_node)
    shortest_length[initial_node.name] = 0
    shortest_path[initial_node.name] = [initial_node.name]


    for node_name in G.vertexes:
        if G.vertexes[node_name] not in processed_nodes:
            unprocessed_nodes.append(G.vertexes[node_name])

    # until all the nodes are processed
    while unprocessed_nodes:

        # record or edges that exist between processed and unprocessed nodes
        edges_to_consider = []

        # record of the shortest edge in the considered edges
        shortest_edge_current_loop = None

        # record of the weight of the said shortest edge
        shortest_edge_weight_current_loop = None

        # get the edges that are eligible conderation
        for edge_name in G.all_edges:
            edge = G.all_edges[edge_name]
            # if the edge can be considered
            if edge.source_vertex in processed_nodes and edge.destination_vertex in unprocessed_nodes:
                edges_to_consider.append(edge)

                # record the shortest edge
                if (shortest_edge_current_loop == None and shortest_edge_weight_current_loop == None) or \
                    edge.weight < shortest_edge_weight_current_loop:
                    
                    shortest_edge_current_loop = edge
                    shortest_edge_weight_current_loop = edge.weight

        # mark the head of the shortest edge as processed
        processed_nodes.append(shortest_edge_current_loop.destination_vertex)
        unprocessed_nodes.remove(shortest_edge_current_loop.destination_vertex)

        # record the weight and the path to this node from the initial node
        shortest_length[shortest_edge_current_loop.destination_vertex.name] = shortest_length[shortest_edge_current_loop.source_vertex.name] + shortest_edge_weight_current_loop
        shortest_path[shortest_edge_current_loop.destination_vertex.name] = shortest_path[shortest_edge_current_loop.source_vertex.name] + [shortest_edge_current_loop.destination_vertex.name]

    print(f"shortest length from : {initial_node}")
    print(shortest_path)
    print(shortest_length)

    return shortest_length
```

```python
# example 
nodes = "s,v,w,t"
edges = "s-v,s-w,v-w,v-t,w-t"
weights = "1,4,2,6,3"
G = create_graph("G",nodes, edges, weights)

dijkstras_shortest_path(G, G.vertexes['s'])
```

```python
# create the graph given for the assignment

# read the file
with open('dijkstrasData.txt') as file:
    nodes = []
    edges = []
    weights = []
    for row in file:
        row = row.split()
        source_node = row[0]
        if source_node not in nodes: nodes.append(source_node)
        for node_weight in row[1:]:
            destination_node = node_weight.split(',')[0]
            weight = node_weight.split(',')[1]

            if destination_node not in nodes: nodes.append(destination_node)

            edge = f"{source_node}-{destination_node}"
            if edge not in edges:
                edges.append(edge)
                weights.append(weight)
    

G = create_graph("G", ",".join(nodes), ",".join(edges), ",".join(weights))

shortest_lenght_dict = dijkstras_shortest_path(G,G.vertexes['1'])

```
